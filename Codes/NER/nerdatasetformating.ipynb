```python
import pandas as pd
import json
import csv
```

```python
with open('/content/drive/MyDrive/Datasets/Bangla-NER-Splitted-Dataset.json') as json_file:
    data = json.load(json_file)


employee_data = data['train']
```

```python
sentence = []
iob_tags = []
bioes_tags = []
```

```python
for emp in employee_data:
  sentence.append(emp['sentence'])
  iob_tags.append(emp['iob_tags'])
  bioes_tags.append(emp['bioes_tags'])
```

```python
len(sentence)
```

```python
lst1=[]
lst2=[]
xlst = []
x = 0
for i in iob_tags:
  x+=1
  for j in i:
    lst1.append(j)
    xlst.append(x)

for i in bioes_tags:
  for j in i:
    lst2.append(j)

lst3=[]
for i in sentence:
  for j in i:
    lst3.append(j)
```

```python
len(lst1)
```

```python
len(lst2)
```

```python
len(lst3)
```

```python
len(xlst)
```

```python
ner = {}
ner['sentence'] = lst3
ner['sid'] = xlst
ner['bioes_tags'] = lst2
ner['iob_tags'] = lst1
```

```python
NERDATA = pd.DataFrame.from_dict(ner)
```

```python
NERDATA.to_csv("/content/train_ner.csv", index=False)
```
