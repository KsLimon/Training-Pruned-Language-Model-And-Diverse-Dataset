# **json to csv**

```python
word=[]
bioes=[]
iob=[]
for emp in train_data:
  for i in emp:
    if i=='sentence':
      for j in emp[i]:
        word.append(j)
    elif i=='bioes_tags':
      for j in emp[i]:
        bioes.append(j)
    elif i=='iob_tags':
      for j in emp[i]:
        iob.append(j)
```

# **json to jsonl**

```python
data_file = open('train.csv', 'w')
 
csv_writer = csv.writer(data_file)
 
count = 0
 
for emp in train_data:
    if count == 0:
 
        header = emp.keys()
        csv_writer.writerow(header)
        count += 1
 
    csv_writer.writerow(emp.values())
 
data_file.close()
```

```python
btags = []
itags = []
for i in range (0, 64155):
  d = {}
  d2 = {}
  dfs = ast.literal_eval(df.sentence[i])
  dfb = ast.literal_eval(df.bioes_tags[i])
  dfi = ast.literal_eval(df.iob_tags[i])
  d["tokens"] = dfs
  d['tags'] = dfb
  d2["tokens"] = dfs
  d2['tags'] = dfi
  btags.append(d)
  itags.append(d2)
```
