```python
import pandas as pd
```

```python
ds_test = pd.read_csv('/content/drive/MyDrive/Shadhu _Kobita/sentiment_test.tsv', encoding='utf-8', header=None, delimiter="\t")
ds_test.columns = ['id', 'text', 'class_label']
ds_test.drop('id', inplace=True, axis=1)
```

```python
ds_test.drop(0, inplace=True, axis=0)
ds_test.head()
```

```python
ds_test['class_label'] = ds_test['class_label'].replace({'Negative':0, 'Positive':1, 'Neutral':2})
```

```python
ds_test.head()
```

```python
labels = list(set(ds_test['class_label']))
print(labels)
```

```python
for i in range(len(labels)):
  ds_test.loc[ds_test['class_label'] == labels[i], 'class_label'] = i
df = ds_test.sample(frac=1) # randomize data
df.head(20)
```

```python
df = df.convert_dtypes(convert_string=True).dropna()
```

```python
list(set(df['class_label']))
```

```python
# split train test
from sklearn.model_selection import train_test_split

train, eval = train_test_split(df,test_size=0.2)
print(len(train))
print(len(eval))
```
