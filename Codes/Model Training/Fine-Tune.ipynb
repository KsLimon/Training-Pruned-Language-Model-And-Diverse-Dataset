# **Dataset**

```python
my_file = open("/content/drive/MyDrive/Shadhu _Kobita/full_corpas.txt", "r",encoding="utf-8")
content_list = my_file.readlines()

import random
random.shuffle(content_list)

from sklearn.model_selection import train_test_split
train, test = train_test_split(content_list, test_size = 0.2)

file1 = open("train.txt","w")
file2 = open("test.txt","w")

for i in train:
    file1.write(i)
file1.close()

for i in test:
    file2.write(i)
file2.close()
```
