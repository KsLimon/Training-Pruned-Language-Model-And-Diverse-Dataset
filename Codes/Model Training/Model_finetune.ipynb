# **Dataset**

```python
my_file = open("/content/xxxxx.txt", "r")
content_list = my_file.readlines()

import random
random.shuffle(content_list)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(content_list, test_size = 0.4)
```

```python
file1 = open("train.txt","w")
file2 = open("test.txt","w")
```

```python
for i in train:
    file1.write(i)
file1.close()

for i in test:
    file2.write(i)
file2.close()
```

# **Training Model**

```python
from simpletransformers.language_modeling import LanguageModelingModel, LanguageModelingArgs

model_args = LanguageModelingArgs()
model_args.vocab_size = 30522
model_args.overwrite_output_dir = True
model_args.reprocess_input_data = True
model_args.num_train_epochs = 3

model = LanguageModelingModel(
    "bert", '/content/drive/MyDrive/Colab Notebooks/ipet01/', use_cuda=True, args=model_args, 
)
# Train the model
# model.train_model("train.txt")

# # Evaluate the model
result = model.eval_model('/content/drive/MyDrive/Colab Notebooks/train1.txt')

# # predictions, raw_outputs = model.predict(["Some arbitary sentence"])
print(result)
```
