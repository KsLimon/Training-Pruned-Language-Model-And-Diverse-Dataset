```python
import pandas as pd
```

```python
train = pd.read_csv("/content/drive/MyDrive/Datasets/Ready_corpas.csv")
train.head()
```

```python
labels = list(set(train['category']))
print(labels)
```

```python
train['category'] = train['category'].replace({'Newspaper':0, 'social_media':1, 'Shadhu':2})
```

```python
train.head()
```

```python
from simpletransformers.classification import ClassificationModel
```

```python
model = ClassificationModel('bert', 'KamrusSamad/tiny_A-2_H-2', num_labels=3, use_cuda=True, args={
    'reprocess_input_data': True,
    'overwrite_output_dir': True,
    'num_train_epochs': 3
})
model.train_model(train)
```
