# **txt formating**

```python
my_file = open("/content/drive/MyDrive/Datasets/full_corpas.txt", "r")
content_list = my_file.readlines()
```

```python
len(content_list)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(content_list, test_size = 0.2)
```

```python
file1 = open("train.txt","w")
```

```python
for i in train:
    file1.write(i)
file1.close()
```

```python
file2 = open("text.txt","w")
```

```python
for i in test:
    file2.write(i)
file2.close()
```
