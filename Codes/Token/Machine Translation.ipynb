```python
with open("/content/drive/MyDrive/Datasets/bnlp-resources/mt/SUPara0.8M/SUPara0.8M_bn.txt") as f:
    train_bn= f.read().splitlines()

with open("/content/drive/MyDrive/Datasets/bnlp-resources/mt/SUPara0.8M/SUPara0.8M_en.txt") as f:
    train_en= f.read().splitlines()

with open("/content/drive/MyDrive/Datasets/bnlp-resources/mt/SUPara0.8M/suparatest2018_bn.txt") as f:
    test_bn= f.read().splitlines()

with open("/content/drive/MyDrive/Datasets/bnlp-resources/mt/SUPara0.8M/suparatest2018_en.txt") as f:
    test_en= f.read().splitlines() 
```

```python
from datasets.dataset_dict import DatasetDict
from datasets import Dataset
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"id": trainid, "translation": traintrans})
train_dataset = Dataset.from_pandas(df)
```

```python
testid = []
testtrans = []
c = 0
for i in range(len(test_bn)):
  dic = {}
  
  dic["bn"] = test_bn[i]
  dic["en"] = test_en[i]
  testid.append(c)
  testtrans.append(dic)
  c+=1
```

```python
df = pd.DataFrame({"id": testid, "translation": testtrans})
test_dataset = Dataset.from_pandas(df)
```

```python
import datasets
row_data = datasets.DatasetDict({"train":train_dataset,"test":test_dataset})
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer
```

```python
tokenizer = AutoTokenizer.from_pretrained("/content/drive/MyDrive/Models/CSE498")
```

```python
source_lang = "bn"
target_lang = "en"
prefix = "translate Bengali to English: "


def preprocess_function(examples):
    inputs = [prefix + example[source_lang] for example in examples["translation"]]
    targets = [example[target_lang] for example in examples["translation"]]
    model_inputs = tokenizer(inputs, max_length=128, truncation=True)

    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=128, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
tokenized_books = row_data.map(preprocess_function, batched=True)
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained('/content/drive/MyDrive/Models/CSE498')
```

```python
from transformers import DataCollatorForSeq2Seq

data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model)
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=1,
)
```

```python
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_books["train"],
    eval_dataset=tokenized_books["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
)
```

```python
trainer.train()
```
