```python
from transformers import AutoTokenizer, AutoModel

bnbert_tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
text = str
list = bnbert_tokenizer.tokenize(text)
```

```python
file1 = open("/content/drive/MyDrive/Shadhu _Kobita/tokens.txt","w",encoding = 'utf-8')
```

```python
for i in list:
    file1.write(i + "\n")
file1.close()
```
